```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Models
from sklearn.ensemble import IsolationForest

df = pd.read_csv('Output.csv', encoding='windows-1252')

df['DATE'] = pd.to_datetime(df['DATE'])    
df['DATE'] = (df['DATE'] - df['DATE'].min())  / np.timedelta64(1,'D')

x= df.values
plt.scatter(x[:, 0], x[:, 1])
```

```python
clf = IsolationForest(contamination=.25)
clf.fit(x)
predictions = clf.predict(x)
print("Anomaly point: ", (predictions < 0).mean())
```

```python
anomalies = np.where(predictions < 0)
plt.scatter(x[:, 0], x[:, 1])
plt.scatter(x[anomalies, 0], x[anomalies, 1], edgecolors='r')
```
